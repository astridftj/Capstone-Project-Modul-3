### **Confusion Matrix Terms:**

1. *TP (True Positive)* : Customers who actually churn and are predicted to churn.

2. *TN (True Negative)* : Customers who actually do not churn and are predicted not to churn.

3. *FP (False Positive)* : Customers who actually do not churn but are predicted to churn.

4. *FN (False Negative)* : Customers who actually churn but are predicted not to churn.

**False Positive**

Consequence: *Operational loss* due to spending promotional costs on the wrong customers, incurring a cost of $10/customer per month.

**False Negative**

Consequence: *Loss of profit* because the customer churns, incurring a cost 5 times higher than the promotional cost to the customer $50/customer per month.

Based on these consequences, I have decided to focus on False Negatives but also not neglect False Positives, with a greater emphasis on recall. Therefore, the primary metric I will use is the **F2-Score**.

The F2-Score is an accuracy measure that considers both precision (the number of true positives divided by the total number of predicted positives) and recall (the number of true positives divided by the total number of actual positives). The F2-Score is the weighted harmonic mean of precision and recall, achieving its optimal value at 1 (perfect precision and recall) and its worst value at 0.

In the F2-Score, recall is considered twice as important as precision. This means it is more critical to capture as many actual positives as possible, even if it means predicting some negatives as positives. This makes the F2-Score highly useful in situations like this, where the cost of False Negatives (customer loss) is higher than the cost of False Positives (unnecessary promotions).

> ## **IQR (Interquartile Range)**

To determine which data points are considered as outliers, we might use statistical measures like the IQR (Interquartile Range). The IQR is calculated as the difference between the 75th percentile (Q3) and 25th percentile (Q1) values of a dataset. Data points that fall below Q1 - 1.5IQR or above Q3 + 1.5IQR could be considered as outliers.

```python
# Dictionary to store outliers' information
outliers_info = {}

# List of numerical columns
numerical_columns = ['Tenure', 'WarehouseToHome', 'NumberOfDeviceRegistered','SatisfactionScore',
       'NumberOfAddress', 'Complain', 'DaySinceLastOrder', 'CashbackAmount',
       'Churn']

for column in numerical_columns:
    # Calculate Q1, Q3, and IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Calculate bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers_condition = (df[column] < lower_bound) | (df[column] > upper_bound)
    outliers = df[outliers_condition]

    # Store information in the dictionary
    outliers_info[column] = {
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'number_of_outliers': outliers_condition.sum(),  # Count of outliers
        'outliers_indices': outliers.index.tolist()  # Indices of outliers
    }

# Summary for outliers_summary
outliers_summary = {key: {'bounds': (value['lower_bound'], value['upper_bound']), 
                          'number_of_outliers': value['number_of_outliers']} 
                    for key, value in outliers_info.items()}

# Creating DataFrame from 'outliers_summary'
outliers_summary_df = pd.DataFrame.from_dict(outliers_summary, orient='index')

# Splitting 'bounds' into 'lothis_bound' and 'upper_bound' columns
outliers_summary_df[['lower_bound', 'upper_bound']] = outliers_summary_df['bounds'].apply(pd.Series)

# Dropping the original 'bounds' column
outliers_summary_df.drop('bounds', axis=1, inplace=True)

# Reordering the columns for a better view
outliers_summary_df = outliers_summary_df[['lower_bound', 'upper_bound', 'number_of_outliers']]

# Display DataFrame as a styled table
styled_table = outliers_summary_df.style.format('{:.0f}')
styled_table
```

> ## **6. Data Preprocessing**

- OneHotEncoder: One-hot encoding is a method used to convert categorical data variables, making them suitable for machine learning algorithms to enhance predictive performance. This process involves transforming each categorical value into a new categorical column and assigning binary values of 1 or 0. Each unique category is represented as a binary vector where all values are zero, except for the index marked with a 1. In this instance, the OneHotEncoder is utilized for the 'MaritalStatus' and 'PreferredOrderCat' columns, both of which are categorical.

- RobustScaler: Robust scaling serves as a preprocessing technique designed to scale features while mitigating the impact of outliers. Unlike MinMax Scaler, which uses the min-max range, robust scaling employs the interquartile range. This scaling algorithm eliminates the median and scales the data based on the quantile range. In this particular scenario, the RobustScaler is applied to the 'Tenure', 'NumberOfDeviceRegistered', 'SatisfactionScore', 'DaySinceLastOrder', and 'CashbackAmount' columns, all of which are numerical.

```python
# Cross validation
models = [logreg, knn, tree, voting, stacking, bagging, rf, adaboost, gboost, xgb]

list_all_f2 = []   # List to store F2 scores from each model and each cross-validation iteration
list_mean_f2 = []  # List to store the mean F2 scores from each model
list_std_f2 = []   # List to store the standard deviation of F2 scores from each model

# F2 score (Because the cost of FN (predicted not churn, but actually churn) is higher than FP (predicted churn, but actually not churn))
f2_score = make_scorer(fbeta_score, beta=2)
f2_score = make_scorer(fbeta_score, beta=2)

# Finding the best model
for i in models:
    # Pipe_model 
    model = Pipeline([
        ('preprocessing', transformer),   # Preprocessing using the transformer
        ('model', i)                      # Modeling using model 'i'
    ])

    cv_score = cross_val_score(
        estimator = model,                # Using the model set in the pipeline
        X = X_train,                      # Feature data for training
        y = y_train,                      # Feature data for training
        cv = 5,                           # Number of folds in cross-validation is 5
        scoring = f2_score,               # Evaluation metric used is f2_score
        n_jobs = -1                       # A parameter used to optimize the use of computational power during cross-validation
    )
    
    list_all_f2.append(cv_score.round(2))
    list_mean_f2.append(cv_score.mean())
    list_std_f2.append(cv_score.std())
```

```python
df_cv = pd.DataFrame()
df_cv['algo'] = models
df_cv['F2_score']= list_all_f2
df_cv['F2_mean'] = list_mean_f2
df_cv['F2_std'] = list_std_f2

df_cv.sort_values(by='F2_mean', ascending=False)
```
